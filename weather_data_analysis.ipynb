```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from glob import glob
import matplotlib.pyplot as plt
```

```python
folder_path = '/Users/mehulchopra/Documents/personal/training/data-analysis-data/weather_data_input/weather_data_set*'
files = glob(folder_path)
print(files)
```

```python
dfs = [pd.read_csv(file, delimiter='|', header=None, names=['station', 'year', 'datemonth', 'hourminute', 'temperature'])
           for file in files]
dfs
```

```python
data = pd.concat(dfs, ignore_index=True)
data
```

```python
data1 = data.dropna()
data1
```

```python
data2 = data.ffill()
data2
```

```python
d3 = data2[['year', 'temperature']]
d3
```

```python
d3.max()
```

```python
d4 = d3.groupby('year')
d4
```

```python
d5 = d4.max()
d5
```

```python
d5.plot(kind='bar')
# plt.show()
plt.savefig('/Users/mehulchopra/Desktop/weather.pdf', format='pdf')
```
